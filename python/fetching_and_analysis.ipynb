# Fetching Historical Data using yfinance
I am using the top 25 aerospace and defense companies from Yahoo Finance.

```python
# import necessary libraries
import numpy as np
import pandas as pd
import yfinance as yf
from itertools import combinations

ticker_list = ['GE', 'RTX', 'BA', 'LMT', 'TDG', 'GD', 'HWM', 'NOC', 'AXON', 'LHX', 'HEI-A', 'RKLB', 'CW', 'TXT', 'WWD', 'BWXT', 'DRS', 'SARO', 'HII', 'KTOS', 'AVAV', 'LOAR', 'MOG-A', 'ACHR', 'HXL']
data = yf.download(ticker_list, period='1y')
```

```python
# create a pandas DataFrame of just the closing prices
data_close = pd.DataFrame(data['Close'])
```
